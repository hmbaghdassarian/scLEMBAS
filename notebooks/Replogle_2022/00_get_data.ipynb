We will download the single-cell data from [Replogle 2022](https://doi.org/10.1016/j.cell.2022.05.013). While scPerturb also has this data, the original paper took special steps for batch-correction during normalization. Thus, we will download the original datastes from [Figshare](https://plus.figshare.com/articles/dataset/_Mapping_information-rich_genotype-phenotype_landscapes_with_genome-scale_Perturb-seq_Replogle_et_al_2022_processed_Perturb-seq_datasets/20029387?file=35775554). 

Specifically, we will get the:
1) normalized single-cell K562 essential genes
2) normalized single-cell rpe1 essential genes

```python
import os
```

```python
data_path = '/nobackup/users/hmbaghda/scLEMBAS/analysis/'
```

Original data:

```python
rpe1_link = 'https://plus.figshare.com/ndownloader/files/35775554'
k562_link = 'https://plus.figshare.com/ndownloader/files/35773075'

links = [rpe1_link, k562_link]
link_names = ['rpe1', 'K562_essential']
cmd = 'wget -O ' + os.path.join(data_path, 'raw/', 'fn') + ' link'
# cmd = 'aria2c -x 16 -s 16 -o ' + os.path.join(data_path, 'raw/', 'fn') + ' link'

for link, fn in zip(links, link_names):
    cmd_ = cmd.replace('link', link)
    cmd_ = cmd_.replace('fn', fn + '_normalized_singlecell_01.h5ad')
    print(cmd_)
```

scPerturb versions:

```python
links = ['https://zenodo.org/record/7041849/files/ReplogleWeissman2022_rpe1.h5ad']
link_names = ['rpe1']

cmd = 'wget -O ' + os.path.join(data_path, 'raw/', 'fn') + ' link'

for link, fn in zip(links, link_names):
    cmd_ = cmd.replace('link', link)
    cmd_ = cmd_.replace('fn', 'scperturb_ReplogleWeissman2022_' + fn + '.h5ad')
    print(cmd_)
```
