Original data from [this paper](https://www.nature.com/articles/nbt.4042#Sec2)

```r
suppressPackageStartupMessages({
    library(Seurat, quietly = T)
    library(SeuratData, quietly = T)
    library(Matrix, quietly = T)
    library(rhdf5, quietly = T)
})
```

```r
data_path = '/nobackup/users/hmbaghda/scLEMBAS/analysis'
```

```r
InstallData("ifnb")
ifnb <- LoadData("ifnb")
```

```r
h5_file <- file.path(data_path, 'raw', 'kang_2017_raw_counts.h5')
sparse_matrix<-ifnb[['RNA']]$counts

# Write the sparse matrix to HDF5 format
h5createFile(h5_file)
h5write(sparse_matrix@x, h5_file, "data")
h5write(sparse_matrix@i, h5_file, "i")
h5write(sparse_matrix@p, h5_file, "p")
h5write(dim(sparse_matrix), h5_file, "dims")

h5write(rownames(sparse_matrix), h5_file, "rownames")
h5write(colnames(sparse_matrix), h5_file, "colnames")

H5close()
```

```r
write.csv(ifnb@meta.data, 
          file.path(data_path, 'raw', 'kan_2017_md.csv'))
```
